```python
auto_vehicles = df[df['Transmission'] == 'Automatic']
num_auto_vehicles = auto_vehicles.shape[0]

if num_auto_vehicles > 0:
    print(f"Yes, the database includes automatic transmission vehicles. There are {num_auto_vehicles} automatic transmission vehicles in the dataset.")
else:
    print("No, the database does not include any automatic transmission vehicles.")
```

```python
single_owner_vehicles = df[df['Owner'] == 0]
num_single_owner_vehicles = single_owner_vehicles.shape[0]

print(f"The number of single person owned vehicles in the dataset: {num_single_owner_vehicles}")
```

```python
df['Depreciation'] = df['Present_Price'] - df['Selling_Price']

# Find the most cost depreciated vehicle
most_depreciated_vehicle = df.loc[df['Depreciation'].idxmax()]

# Find the least cost depreciated vehicle
least_depreciated_vehicle = df.loc[df['Depreciation'].idxmin()]

print("Most cost depreciated vehicle:")
print(most_depreciated_vehicle[['Car_Name', 'Present_Price', 'Selling_Price', 'Depreciation']])

print("\nLeast cost depreciated vehicle:")
print(least_depreciated_vehicle[['Car_Name', 'Present_Price', 'Selling_Price', 'Depreciation']])
```

```python
# List of keywords to identify bike names
bike_keywords = ["Royal Enfield", "KTM", "Bajaj", "Honda", "Yamaha","Activa", "Hero", "TVS", "Suzuki", "Hyosung", "UM", "Mahindra"]

# Function to check if a name contains any bike keyword
def is_bike(name):
    return any(keyword in name for keyword in bike_keywords)

# Create a new column to identify bikes
df['Is_Bike'] = df['Car_Name'].apply(is_bike)

# Separate DataFrames for cars and bikes
bikes_df = df[df['Is_Bike'] == True].drop(columns=['Is_Bike'])
cars_df = df[df['Is_Bike'] == False].drop(columns=['Is_Bike'])

# Display the first few rows of each DataFrame
print("Bikes DataFrame:")
print(bikes_df.head())

print("\nCars DataFrame:")
print(cars_df.head())

# Display the last few rows of each DataFrame
print("\nLast few rows of Bikes DataFrame:")
print(bikes_df.tail())

print("\nLast few rows of Cars DataFrame:")
print(cars_df.tail())

# Display a summary of each DataFrame
print("\nSummary of Bikes DataFrame:")
bikes_df.info()

print("\nSummary of Cars DataFrame:")
cars_df.info()

# Display basic statistics of each DataFrame
print("\nBasic statistics of Bikes DataFrame:")
print(bikes_df.describe())

print("\nBasic statistics of Cars DataFrame:")
print(cars_df.describe())
```
